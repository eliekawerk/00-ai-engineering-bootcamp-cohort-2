# Imports

```python
from langsmith import Client
```

## Attach Feedback to specific trace in Langsmith

```python
ls_client = Client()
```

```python
ls_client.create_feedback(
    run_id="dcbd8646-a291-410e-b1b9-2ff062566668",
    key="feedback",
    score=0,
    feedback_source_type="api",
)
```

```python
ls_client.create_feedback(
    run_id="dcbd8646-a291-410e-b1b9-2ff062566668",
    key="feedback1",
    score=1,
    feedback_source_type="api",
)
```

```python
ls_client.create_feedback(
    run_id="dcbd8646-a291-410e-b1b9-2ff062566668",
    key="comment",
    value="this is a long form feedback, you should use to improve the system",
    feedback_source_type="api",
)
```
